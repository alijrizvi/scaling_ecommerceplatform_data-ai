```python
# Synthetic Dataset for SQL Querying with the main DataFrame

from pyspark.sql import functions as F, types as T

data = [
    ("2010-05-08", 25.0, 6.0, "Amitabh Bachchan"),
    ("2017-07-01", None, 9.0, "Amitabh Bachchan"),
    ("2017-07-05", 6.0, None, "Rishi Kapoor"),
    ("2011-08-11", None, 7.0, None),
    ("2012-10-11", 28.0, None, "Amitabh Bachchan"),
    ("2013-11-15", None, None, "Govinda"),
    ("2014-09-12", None, None, None),
    ("2015-04-14", 30.1, 8.2, "Brad Pitt"),
    ("2017-12-01", 40.0, 12.0, "Brad Pitt")
]

schema = "releaseday string, temperature double, windspeed double, staractor string"
movies2 = spark.createDataFrame(data, schema)
movies2 = movies2.withColumn("releaseday", F.to_date("releaseday", "yyyy-MM-dd")) # Normalize to Date Type

display(movies2)
```

```python
# To Run SQL Queries on this DataFrame, we need to register it as a Temporary Table

movies2.createOrReplaceTempView("movies2") # Session-scoped; Valid only for this Current Session and Notebook
movies2.createGlobalTempView("global_movies2") # Should be on the Same Cluster but on any other Notebook
```

```python
%sql

SELECT staractor, ROUND(AVG(temperature), 1) AS avg_temp
FROM movies2
GROUP BY staractor
ORDER BY staractor DESC;  -- Average Temperature on Day of Release of Movie(s) of each Star Actor
```

```python
spark.sql("""
SELECT releaseday, temperature, windspeed, staractor
FROM movies2
WHERE temperature IS NOT NULL
ORDER BY releaseday
""").show()
```

```python
# Performing JOINs in Spark

display(movies2.join(movies, movies2.staractor == movies.Star1, how = "left_anti")) # Left-Anti Join (All non-matching Actors in "movies2")
```
